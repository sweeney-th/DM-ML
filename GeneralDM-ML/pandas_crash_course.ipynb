# Missing Values

```python
# check that you don't have any by asserting the number of rows is the number of rows that don't have missing data
#assert len(digits.isnull()) == len(digits)
```

If the exist, find them using isnull()

Note: is null will pick up on NaN, as we see in First:A

```python
# set value at index 0 of column 'a' to NaN
df.at['First', 'A'] = np.nan

# look for them 
df.isnull()
```

To demonstrate dropping, we will make a second data frame and remove some of the data from it.

NOTE: copies must be made explicitly to keep the original dataframe from being modified.

```python
# make a copy 
df2 = df.copy()

# set some to NaN
df2.at["First",'B']  = 'NaN'
df2.at["Second",'B'] = 'NaN'
df2.at["First", 'C'] = 'NaN'
df2
```

Drop rows where ALL values are missing:

```python
# call dropna on axis 0 (rows) - "First" should be gone 
drop_any_missing = df2.dropna(axis = 0, how = 'all')
drop_any_missing
```

Notice that 'First' is gone because it was NaN in both places. 'Second' is still there are as only one of the values is 'NaN'

```python
# any missing gets dropped. "First" and "Second" should be gone 
drop_any_missing = df2.dropna(axis = 0, how = "any")
drop_any_missing
```

This will do the same this to columns if you use axis 1

```python
df3 = df.copy()

# omit the .at() to add empty column, just like R
df3['C'] = np.nan
df3
```

```python
# new df is df3 dropping all columns that are all NaN - A is kept (not all NaN) but C is not 
columns_drop_all = df3.dropna(axis = 1, how = "all")
columns_drop_all
```

```python
# now with any NaN 
columns_drop_any = df3.dropna(axis = 1, how = "any")
columns_drop_any
```

This leaves us with just B. A had one missing value and C was only missing values. 

# Imputation via the Mean 

```python
df2.head()
```

Use the "mean" function from Numpy to get the mean, and then the "fillna" method from Pandas to put the mean of the column where missing values are 

```python
# get the mean using numpy
mean = np.mean(df2)

# print it so we can see
print(mean)

# works 'magically' not specifying columns 
df_interpolated = df2.fillna(mean)

df_interpolated.head()
```

### The interpolate() Method

The interpolate method fills np.nans with interpolated data automatically! Linear methods are the default but tons are available. Very cool. 

It has some limits built in so you must choose to override them. For instance, you can set how many instances can be interpolated, and you must specify if you want to to add values forward and backwards. 

https://pandas-docs.github.io/pandas-docs-travis/missing_data.html

https://stackoverflow.com/questions/31332981/pandas-interpolation-replacing-nans-after-the-last-data-point-but-not-before-th

```python
# confirm there are still missing values 
df2.head()
```

We will use the linear method to calculate a value for missing pieces 

```python
# call the method with lr, and let it work in both directions 
df2.interpolate(method = "linear", limit_direction = "both")
```

# Applying a Function to DataFrames

```python
function_df = df.copy()
function_df
```

```python
# define a function that acts on X increasing it by 100
increase_by_100 = lambda x: (x + 100)
```

### Apply to the whole dataframe

```python
# apply to the whole dataframe
function_df = function_df.apply(increase_by_100)
function_df
```

### Apply to Specific Rows/Columns 
Use map to replace a column with the function mapped version 

Note that the second row will have been increased by 100 twice 

```python
# replace a row with the same thing but with the function mapped
function_df.loc["Second"] = function_df.loc["Second"].map(increase_by_100)
function_df
```

After applying it to the "E" columns, all of those values will have been increase by 100 twice, except for Second:E, which will have been increased by 100 three times: once as a member of the dataframe, once as a member of "Second", once as a member of "E".

```python
# do the same for a column 
function_df["E"] = function_df["E"].map(increase_by_100)
function_df
```
